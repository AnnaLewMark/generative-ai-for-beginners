 2. Анализ таблицы для определения форматов значений в ячейках по столбцам. Вывод сообщения о том, какие имена столбцов содержат данные разных форматов, и указания, нужно ли исправлять форматы. Просмотр первых 7 строк таблицы.

```python
import pandas as pd  # Импорт библиотеки Pandas для работы с табличными данными

# Загрузка файла из директории среды
file_path = '/content/combined_orders.csv'

# Чтение файла с указанием символа ';' в качестве разделителя
data = pd.read_csv(file_path, delimiter=';', encoding='utf-8-sig')

# Отображение первых 7 строк таблицы
display(data.head(7))

# Анализ форматов данных по столбцам
columns_with_mixed_types = []  # Список для хранения имен столбцов с разными типами данных
for column in data.columns:  # Проходим по каждому столбцу в данных
    unique_types = set(data[column].apply(type))  # Определяем уникальные типы данных в столбце
    if len(unique_types) > 1:  # Если в столбце более одного типа данных
        columns_with_mixed_types.append(column)  # Добавляем имя столбца в список

# Выводим результаты анализа
if columns_with_mixed_types:  # Если есть столбцы с разными типами данных
    print('Столбцы с данными разных форматов:', ', '.join(columns_with_mixed_types))
    print('Требуется корректировка форматов.')
else:  # Если все столбцы содержат данные одного типа
    print('Все столбцы содержат данные одного формата. Корректировка не требуется.')
```

3. Вывод названий всех столбцов и типов форматов данных в них.

```python

# Функция для анализа форматов значений в столбцах
def analyze_and_display_column_formats(df):
    # Проходим по каждому столбцу в DataFrame
    for column in df.columns:
        # Применяем set для определения уникальных типов данных в столбце
        types = set(df[column].dropna().apply(lambda x: type(x).__name__))
        # Вывод названия столбца и уникальных типов данных
        print(f"{column}: {', '.join(types)}")

# Вызов функции анализа
analyze_and_display_column_formats(data)
```

4. Поиск строк с пустыми значениями одновременно в столбцах "Товар",	"Название товара",	"Название категории",	"Ссылка на изображение",	"Стоимость товара",	"Количество".Вывод на экран общего количества таких строк.

```python
print(f"Количество строк всего: {data.shape[0]}")

# Список столбцов, в которых мы ищем пустые значения
columns_to_check = [
    "Товар", "Название товара", "Название категории",
    "Ссылка на изображение", "Стоимость товара", "Количество"
]

# Поиск строк, где все указанные столбцы содержат пустые значения
empty_rows = data[columns_to_check].isna().all(axis=1)

# Подсчет и вывод общего количества таких строк
total_empty_rows = empty_rows.sum()
print(f"Общее количество строк с пустыми значениями в указанных столбцах: {total_empty_rows}")
```

5. Копирование информации о заказе и заказчике вниз по каждому товару в заказе.

```python
import numpy as np

# Столбцы, содержащие информацию о заказе и заказчике
order_info_cols = ["№", "Оформлен", "Заказчик", "Пол", "Тип заказа", "Время доставки", "Стоимость", "Статус заказа"]

# Инициализация переменной для хранения текущей информации о заказе
current_order_info = None

# Итерация по строкам DataFrame
for index, row in data.iterrows():
    if pd.notna(row["№"]):  # Если в текущей строке есть номер заказа, обновляем информацию о заказе
        current_order_info = row[order_info_cols]
    else:  # Если номер заказа отсутствует, значит это строка товара, и мы копируем в неё информацию о заказе
        data.loc[index, order_info_cols] = current_order_info

# Отображаем результат
display(data.head(20))
```

6. Удаление всех строк с одновременно пустыми ячейками в столбцах с информацией о товарах.

```python
# Создаем копию данных
data_copy = data.copy()

# Сохраняем копию данных в новый файл
data_copy.to_csv('/content/new_dataset.csv', sep=';', index=False, encoding='utf-8-sig')

# Считаем количество строк всего
print(f"Количество строк всего: {data_copy.shape[0]}")

# Список столбцов, в которых мы ищем пустые значения
columns_to_check = [
    "Товар", "Название товара", "Название категории",
    "Ссылка на изображение", "Стоимость товара", "Количество"
]

# Поиск строк, где все указанные столбцы содержат пустые значения
empty_rows = data_copy[columns_to_check].isna().all(axis=1)

# Подсчет и вывод общего количества таких строк
total_empty_rows = empty_rows.sum()
print(f"Общее количество строк с пустыми значениями в указанных столбцах: {total_empty_rows}")

# Удаление строк, где все указанные столбцы пусты
data_cleaned = data_copy.loc[~empty_rows]

# Сброс индекса после удаления строк
data_cleaned.reset_index(drop=True, inplace=True)

# Проверка количества оставшихся строк после удаления пустых
print(f"Количество строк после удаления: {data_cleaned.shape[0]}")

# Вывод первых 10 строк очищенных данных
display(data_cleaned.head(10))

# Сохраняем очищенные данные в тот же файл
data_cleaned.to_csv('/content/new_dataset.csv', sep=';', index=False, encoding='utf-8-sig')
```

7. Удаление строк с отмененными заказами.

```python
# Поиск всех строк, где в столбце "Статус заказа" значение ячейки "Отменен"
cancelled_orders = data_cleaned[data_cleaned['Статус заказа'] == 'Отменен']

# Выводим количество найденных строк с "Отменен"
print(f"Количество отмененных заказов: {cancelled_orders.shape[0]}")

# Удаление всех строк, где в столбце "Статус заказа" значение ячейки "Отменен"
data_cleaned = data_cleaned[data_cleaned['Статус заказа'] != 'Отменен']

# Выводим информацию об окончании удаления строк с отмененными заказами
print("Строки с отмененными заказами удалены.")

# Выводим информацию о количестве всех строк в файле после удаления строк с отмененными заказами
print(f"Количество всех строк в файле после удаления: {data_cleaned.shape[0]}")

# Сохраняем изменения в файл
data_cleaned.to_csv('/content/new_dataset.csv', sep=';', index=False, encoding='utf-8-sig')
```

8. Заполнение значений Время доставки значениями из Оформлен для заказов со статусом Самовывоз

```python
# Поиск строк с пустыми значениями в столбце "Время доставки"
empty_delivery_time_rows = data_cleaned[data_cleaned['Время доставки'].isna()]

# Вывод количества таких строк
print(f"Количество строк с пустым 'Время доставки': {empty_delivery_time_rows.shape[0]}")

# Заполнение пустых значений в "Время доставки" значениями из "Оформлен"
data_cleaned.loc[data_cleaned['Время доставки'].isna(), 'Время доставки'] = data_cleaned['Оформлен']

# Вывод на экран 5 строк, где тип заказа "Самовывоз"
display(data_cleaned[data_cleaned['Тип заказа'] == 'Самовывоз'].head(5))

# Сохранение изменений в файл
data_cleaned.to_csv('/content/new_dataset.csv', sep=';', index=False, encoding='utf-8-sig')
```

9. Замена формата значений в столбцах "№" и "Количество" на формат целых чисел.

```python
# Преобразование формата значений в столбцах '№' и 'Количество' на формат целых чисел
data_cleaned['№'] = data_cleaned['№'].astype('Int64')
data_cleaned['Количество'] = data_cleaned['Количество'].astype('Int64')

# Сохранение изменений в файл
data_cleaned.to_csv('/content/new_dataset.csv', sep=';', index=False, encoding='utf-8-sig')

# Проверяем результат, отображая первые несколько строк
display(data_cleaned.head())
```
